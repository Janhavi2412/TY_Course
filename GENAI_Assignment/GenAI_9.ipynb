```python
from moviepy.editor import *

# Create a text clip
text_clip = TextClip("Hello, AI Animation!", fontsize=70, color='white')
text_clip = text_clip.set_duration(5).set_position('center')

# Create a background clip
bg_clip = ColorClip(size=(1280, 720), color=(0, 0, 255), duration=5)

# Overlay text on background
final_clip = CompositeVideoClip([bg_clip, text_clip])

# Save the video
final_clip.write_videofile("simple_animation.mp4", fps=24)
```

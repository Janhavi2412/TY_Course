```python
# Install dependencies
!pip install torch torchaudio # useful for audio processing
!pip install TTS # for text to speech task

import torch
from TTS.api import TTS
import IPython.display as ipd

def setup_tts_model():
    """Load a pre-trained Tacotron or WaveNet model."""
    model_name = "tts_models/en/ljspeech/tacotron2-DDC"  # Tacotron2 model
    tts = TTS(model_name=model_name).to("cuda" if torch.cuda.is_available() else "cpu")
    return tts

def text_to_speech(text, tts, output_path="output.wav"):
    """Generate speech from text using the pre-trained model."""
    tts.tts_to_file(text=text, file_path=output_path)
    return output_path

# Example Usage
tts_model = setup_tts_model()
text = "Hello , welcome you all in the world of Generative AI"
audio_file = text_to_speech(text, tts_model)

# Play the generated speech
ipd.display(ipd.Audio(audio_file))
```
